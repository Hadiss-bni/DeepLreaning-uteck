```python
import cv2
import numpy as np
print(cv2.__version__)
```

```python
image = cv2.imread(r'C:\Users\asus\Desktop\file-20200803-24-50u91u.jpg')
print(image.shape)
```

```python
cv2.imshow('Original', image)

#for blurring the face:
kernel_7x7 = np.ones((7, 7), np.float32) / 49
blurred = cv2.filter2D(image[205:410,308:616,:], -1, kernel_7x7)
cv2.imshow('Blured-image', blurred)

#for drawing a rectangle on the face:
rectangle_image = cv2.rectangle(image, (308,205), (616,410), (100,0,100), -1)
cv2.imshow("Rectangle", rectangle_image)

#or you can uncomment the code below for the same task:
#image[205:410,308:616,:]=0
#cv2.imshow('It is a black cat!', image)



cv2.waitKey()
cv2.destroyAllWindows()
```
